## Struttura di un figura Matplot:d

una figura Matplot é divisa in :

- Figure: é  la figura intera che contiene una o più **axes** (grafici). Lo si può pensare come un canvas che contiene grafici.
- Axes: é un grafico, ogni Axes contiene un titolo, una x-label e una y-label.
- Axis:  sono gli assi del grafico
- Artist: tutto quello che si può vedere nella figura.

# plt.plot()
con `plt.plot()` puoi disegnare qualsiasi grafico.

specificando i due argomenti X e Y per i dati, e oltre a questi due un terzo argomento in forma di stringa che indica il colore e il tipo di linea. Di default é il    `b-` che indica una linea blu solida, ci possono essere altre colorazioni e stile di tratto naturalmente:

```python
plt.figure(figsize=(10,4))
plt.plot([1,2,423,23], [2,32,42,42],'-|k')
```

# plt.subplot():

possiamo utilizzare `subplot()` per poter aggiungere piú plot alla stessa figura. Accetta tre argomenti:

1. nrows = numero di righe
2. ncols = numero di colonne
3. index = numero del subpot

```python
plt.subplot(1,3,1) # una riga  e tre  colonne
plt.plot([2,1,3,4], [1,2,4,5],'r^')

plt.subplot(1,3,3)
plt.plot(x_1,'go')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3, figsize=(10,5))
axes[0].plot([2,1,3,4], [1,2,4,5],'r^')
axes[1].plot(x_1,'go')
axes[2].plot(x_3,'ko')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'r', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'b', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k', label='three')


# Set Legend
ax.legend(loc='best')

# Set Title
ax.set_title('matplotlib plot motherfucker!!')

# Set X and Y axis label
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
```

```python
df_mesi = pd.read_csv("google_trands/mesi.csv",sep=';')
df_30gg = pd.read_csv("google_trands/30_giorni.csv")
df_7gg = pd.read_csv("google_trands/7_giorni.csv")
df_24h = pd.read_csv("google_trands/24_ore.csv")
df_mappe = pd.read_csv("google_trands/geoMap.csv",sep=',')
```

```python
df_mesi['Mese'].count()
```

```python
fig, ((ax1, ax2), (ax3,ax4))= plt.subplots(2,2,figsize=(9,10))

ax1.plot(df_mesi['Mese'],df_mesi['napoli campione: (Italia)'],'b',label = "Mesi")
ax1.set_xlim([min(df_mesi["Mese"]), max(df_mesi["Mese"])])
ax1.set_xscale('linear')
ax1.set_title('20 anni')
ax1.set_xlabel('Tempo')

ax2.plot(df_30gg['Giorno'],df_30gg['napoli campione: (Italia)'],'b',label = "30gg")
ax2.set_title('30 giorni')
ax2.set_xlim([min(df_30gg["Giorno"]), max(df_30gg["Giorno"])])
ax2.set_xscale('linear')
ax2.set_xlabel('Tempo')

ax3.plot(df_7gg['Tempo'],df_7gg['napoli campione: (Italia)'],'b',label = "7gg")
ax3.set_title('7 giorni')
ax3.set_xscale('linear')
ax3.set_yscale('linear')
ax3.set_xlabel('Tempo')

ax4.plot(df_24h['Tempo'],df_24h['napoli campione: (Italia)'],'b',label = "24h")
ax4.set_title('24 ore')
ax4.set_xscale('linear')
ax4.set_xlabel('Tempo')

fig.suptitle('andamento del numero di  ricerche su google di "napoli campione"',fontsize=17,font='Helvetica')
```
